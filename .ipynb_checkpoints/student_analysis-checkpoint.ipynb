# Student Performance Data Analysis

# Objective
The objective of this project is to analyze the student performance dataset using Python. The dataset will be cleaned, explored, analyzed and visualized to gain insights into student academic performance.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("student-mat.csv")
df.head()
```

```python
print(df.shape)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

## Data Cleaning
The dataset contains 649 records and 33 attributes.

### Missing Values
No missing values were found in any column.

### Duplicate Records
No duplicate records were found in any column.

Therefore, no additional cleaning was required.

```python
print("Average Final Grade: ", df['G3'].mean())
```

## Analysis Question 1

The Average Final Grade (G3) was calculated to understand the overall academic performance of students.

```python
above_15 = df[df['G3'] > 15]
print("Students scoring above 15: ", len(above_15))
```

## Analysis Question 2

This analysis identifies high-performing students with final grades greater than 15.

```python
correlation = df['studytime'].corr(df['G3'])
print("Correlation between studytime and G3: ", correlation)
```

## Analysis Question 3

Correlation analysis is used to determine whether increased study time is associated with better academic performance.

```python
gender_avg = df.groupby('sex')['G3'].mean()
print(gender_avg)
```

```python
plt.figure(figsize = (8, 5))
plt.hist(df['G3'], bins = 10)
plt.title("Distribution of final grades: ")
plt.xlabel("Grade")
plt.ylabel("Frequency")
plt.show()
```

### Histogram

Shows how grades are distributed.

```python
plt.figure(figsize = (8, 5))
plt.scatter(df['studytime'], df['G3'])
plt.title("Study time vs Final Grade")
plt.xlabel("Study time")
plt.ylabel("Final Grade")
plt.show()
```

### Scatter plot

Visualizes the relationship between study time and grades.

```python
gender_avg.plot(kind = 'bar')
plt.title("Average Grade by Gender")
plt.xlabel("Gender")
plt.ylabel("Average G3")
plt.show()
```

### Bar Chart

Compares average grades of male and female students.
